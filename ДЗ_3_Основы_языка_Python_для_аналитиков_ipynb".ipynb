###1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома. Если реновации дома не было, то в новом признаке поставьте 0

```python
df['delta_renovated'] = df['yr_renovated']-df['yr_built']
#отталкиваясь от того , что если реновации не было , то разница отрицательная
df.loc[df['delta_renovated'] <= 0, 'delta_renovated'] = 0 # указываем с использование loc, обращаемся к столбцу delta_renovated и пишем если оно меньше или равно нулю, то пишем ноль



```

```python
df['delta_renovated'] = [0 if x <= 0 else x for x in df['delta_renovated']] # альтернативный способ обращаемся также к новому признаку созданному и говорим , что присвоим ноль, если меньше либо равно нулю в этом признаке
```

```python
df['delta_renovated'] = df['delta_renovated'].apply(lambda x: 0 if x <= 0 else x)
 # способ через лямбда-функцию к каждому элементу столбца delta_renovated. Если значение x меньше или равно 0, то оно заменяется на 0, иначе остается без изменений. Результат присваивается обратно в столбец delta_renovated в DataFrame.
```
