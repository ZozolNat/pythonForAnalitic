###2.1. Изучите распределение домов от наличия вида на набережную

```python
Waterfront_df = df['waterfront'].value_counts() # приравниваем/наываем иначе
Waterfront_df.head()
```

```python
plt.figure(figsize=(10, 8))
plt.pie(Waterfront_df.values, labels=Waterfront_df.index, autopct='%.1f%%');
plt.title('Дома с видом на набрежную')
plt.legend(Waterfront_df.index, loc='upper right');
```

Из диаграммы видно, что колличество домов с видом на набержную менее 1 процента.

В) Взаимосвязь стоимости дома и колличества ванных

```python
plt.figure(figsize=(10,4))
sns.boxplot(x=df["bathrooms"], y=df["price"])
plt.xlabel("bathrooms")
plt.ylabel("price");
```

Анализируем количество ванных комнат на стоимость недвижимости.(коробчатая диаграмма, в которой переменная «ванные комнаты» откладывается по оси X, а переменная «цена» — по оси Y.) Г) Взаимосвязь стоимости дома и общей площади

```python
plt.figure(figsize=(10,4))
sns.jointplot(x=df["price"], y=df["sqft_above"], kind='reg')
plt.ylabel("sqft_above")
plt.xlabel("price");
```

Мы видим точечный график распределения взаимного рапределения и можем наблюдать, что чем больше стоимость домов (price) тем больше общая площадь дома (sqft_above). Значит подтверждается гипотеза, что чем больше площадь тем больше стоимость.

Д) Влияния индекса( а значит какого-то района или горда) на стоимость жилья.

```python
plt.figure(figsize=(10,4))
sns.jointplot(x=df["price"], y=df["zipcode"], kind='kde')
plt.ylabel("zipcode")
plt.xlabel("price");
```

Исходя из графика видно, что стоимость домов в городе с индексом 98100 и 98000 выше. Можно предположить, что это более дорогие /элитные города.
